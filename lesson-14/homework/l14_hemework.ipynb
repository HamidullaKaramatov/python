```python
import json


def read_students_json(file_name):
    try:
     
        with open(file_name, 'r') as file:
            data = json.load(file)

        
        for student in data:
            print(f"Talaba ID: {student.get('id')}")
            print(f"Ism: {student.get('name')}")
            print(f"Yosh: {student.get('age')}")
            print(f"Bahosi: {student.get('grade')}")
            print("-" * 20) 

    except FileNotFoundError:
        print(f"{file_name} fayli topilmadi.")
    except json.JSONDecodeError:
        print(f"{file_name} faylidan JSON ma'lumotlarini o'qishda xato.")
    

file_name = 'talabalar.json'


read_students_json(file_name)
```

```python
#3
import json


BOOKS_FILE = 'books.json'


def read_books():
    try:
        with open(BOOKS_FILE, 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        return []  


def write_books(books):
    with open(BOOKS_FILE, 'w') as file:
        json.dump(books, file, indent=4)

# Yangi kitob qo'shish 
def add_book():
    books = read_books()
    new_book = {}
    
    new_book["id"] = len(books) + 1  
    new_book["title"] = input("Kitob nomini kiriting: ")
    new_book["author"] = input("Muallifni kiriting: ")
    new_book["year"] = int(input("Chop etilgan yilni kiriting: "))
    
    books.append(new_book)
    write_books(books)
    print("Yangi kitob muvaffaqiyatli qo'shildi!")

# Mavjud kitobni yangilash 
def update_book():
    books = read_books()
    book_id = int(input("Yangilamoqchi bo'lgan kitobning ID raqamini kiriting: "))
    
    # ID bo'yicha kitobni topish
    book = next((b for b in books if b['id'] == book_id), None)
    
    if book:
        book["title"] = input(f"Yangi nomni kiriting (hozirgi: {book['title']}): ")
        book["author"] = input(f"Yangi muallifni kiriting (hozirgi: {book['author']}): ")
        book["year"] = int(input(f"Yangi yilni kiriting (hozirgi: {book['year']}): "))
        
        write_books(books)
        print("Kitob muvaffaqiyatli yangilandi!")
    else:
        print("Berilgan ID ga ega kitob topilmadi.")

# Kitobni o'chirish 
def delete_book():
    books = read_books()
    book_id = int(input("O'chirmoqchi bo'lgan kitobning ID raqamini kiriting: "))
    
    #  ID bo'yicha topib o'chirish
    books = [b for b in books if b['id'] != book_id]
    
    write_books(books)
    print("Kitob muvaffaqiyatli o'chirildi!")

# Menyu va foydalanuvchi tanlovlarini boshqarish
def menu():
    while True:
        print("\n--- Kitoblarni boshqarish ---")
        print("1. Yangi kitob qo'shish")
        print("2. Mavjud kitobni yangilash")
        print("3. Kitobni o'chirish")
        print("4. Chiqish")
        
        choice = input("Tanlovni kiriting: ")
        
        if choice == '1':
            add_book()
        elif choice == '2':
            update_book()
        elif choice == '3':
            delete_book()
        elif choice == '4':
            print("Xayr!")
            break
        else:
            print("Noto'g'ri tanlov, iltimos qaytadan kiriting.")

if __name__ == "__main__":
    menu()
```

```python
import requests
import random

# OMDb API URL va API kalitini o'rnatish
OMDB_API_URL = "http://www.omdbapi.com/"
API_KEY = "http://www.omdbapi.com/?i=tt3896198&apikey=d24b71c1"  # O'z API kalitini shu yerga qo'ying

# Janr bo'yicha filmlarni olish funksiyasi
def get_movies_by_genre(genre):
    try:
        params = {
            'apikey': API_KEY,
            'type': 'movie',
            'genre': genre,
            's': genre  # Search query is the genre itself
        }
        response = requests.get(OMDB_API_URL, params=params)
        
        if response.status_code == 200:
            data = response.json()
            if data['Response'] == 'True':
                return data['Search']
            else:
                print(f"No movies found for the genre: {genre}. Please try another genre.")
                return []
        else:
            print(f"Failed to fetch data. Status Code: {response.status_code}")
            return []
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return []

# Tanlangan filmdan tasvir va boshqa tafsilotlarni olish
def get_movie_details(imdb_id):
    params = {
        'apikey': API_KEY,
        'i': imdb_id
    }
    
    response = requests.get(OMDB_API_URL, params=params)
    
    if response.status_code == 200:
        data = response.json()
        if data['Response'] == 'True':
            return {
                'plot': data.get('Plot', 'No plot available'),
                'actors': data.get('Actors', 'No actors listed'),
                'rating': data.get('imdbRating', 'No rating available'),
                'poster': data.get('Poster', 'No poster available')
            }
    return {}

# Tasodifiy filmni tavsiya qilish funksiyasi
def recommend_movie(movies):
    if not movies:
        return None
    random_movie = random.choice(movies)
    movie_title = random_movie['Title']
    movie_year = random_movie['Year']
    movie_imdb_id = random_movie['imdbID']
    
    # Tanlangan filmni batafsil olish
    movie_details = get_movie_details(movie_imdb_id)
    
    return movie_title, movie_year, movie_details

# Asosiy funksiya
def main():
    print("Welcome to the Movie Recommendation System!")
    genre = input("Please enter a movie genre (e.g., Action, Comedy, Drama): ").strip().capitalize()
    
    # Janr bo'yicha filmlar olish
    movies = get_movies_by_genre(genre)
    
    # Tasodifiy filmni tavsiya qilish
    if movies:
        movie_title, movie_year, movie_details = recommend_movie(movies)
        print("\nRecommended Movie:")
        print(f"Title: {movie_title} ({movie_year})")
        print(f"Plot: {movie_details['plot']}")
        print(f"Actors: {movie_details['actors']}")
        print(f"IMDb Rating: {movie_details['rating']}")
        print(f"Poster URL: {movie_details['poster']}")
    else:
        print("No movies found for this genre. Please try again with a different genre.")

if __name__ == "__main__":
    main()
```
